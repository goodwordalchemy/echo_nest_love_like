###Silhouette Analysis 
*  for finding the optimal k in a kmeans clustering. silhouette plot displays a measure of how close each point in one cluster is to points in neighboring clusters
*  silhouette coefficients [-1,1]
    *  +1 indicating that all 
    *  0 indicates that the sample is on or very close to the decision boundary between two neighboring clusters.
    *  negative values indicate that the point may have been assigned to the wrong cluster.

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

range_n_clusters = range(2,15)
silhouette_avg_list = []
dataset = num_data[data.love==1]

#svd for visualization
svd = TruncatedSVD(n_components=2)
X_2d = svd.fit_transform(dataset)

for n_clusters in range_n_clusters:
    fig, (ax1, ax2) = plt.subplots(1,2)
    fig.set_size_inches(18, 7)
    
    #silhouette plot
    ax1.set_xlim([-.1, .5])
    ax1.set_xticks(np.arange(-.1, .5, .5))
    ax1.set_ylim([0, len(dataset) + (n_clusters+1)*10])
    
    clusterer = KMeans(n_clusters=n_clusters)
    cluster_labels = clusterer.fit_predict(dataset)
    
    silhouette_avg = silhouette_score(dataset, cluster_labels)
    silhouette_avg_list.append((n_clusters, silhouette_avg))
    print "for n_clusters=%d, "%n_clusters + \
          "The average silhouette_score is: %.3f"%silhouette_avg
    
    sample_silhouette_values = silhouette_samples(dataset, cluster_labels)
    y_lower = 10
    for i in range(n_clusters):
        #aggregate scores for cluster and sort them
        ith_cluster_silhouette_values = \
                sample_silhouette_values[cluster_labels == i]
            
        ith_cluster_silhouette_values.sort()
        
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
        
        color = cm.spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                         facecolor=color, edgecolor=color, alpha=0.7)
        
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10
    
    ax1.set_title("Silhouette plot for various clusters.")
    ax1.set_xlabel("silhouette coefficient values")
    ax1.set_ylabel("Cluster Label")
    
    # clustering plot
    colors = cm.spectral(cluster_labels.astype(float)/n_clusters)
    
    ax2.scatter(X_2d[:, 1], X_2d[:,0], marker='.',
                s=40, lw=0, alpha=0.7)
    centers = clusterer.cluster_centers_
    trans_centers = svd.fit_transform(centers)
    ax2.scatter(trans_centers[:, 1], trans_centers[:, 0],
                marker='o', c='white', alpha=1, s=200)
    ax2.set_title("Visualization of clustered data")
    ax2.set_ylabel('PC1')
    ax2.set_xlabel('PC2')
    fig.tight_layout()
    plt.subplots_adjust(top=.8)
    plt.suptitle('Silhouette analysis for KMeans clustering '+
                  'on music dataset with n_clusters %d'%n_clusters,
                 fontsize=14, fontweight='bold')
    plt.show()
```

```python
n_clusters, silhouette_scores = zip(*silhouette_avg_list)
fig, ax = plt.subplots()
ax.scatter(n_clusters, silhouette_scores)
ax.set_title('Silhouette Scores for Various K Values on Playlist Dataset')
ax.set_xlabel('K clusters')
ax.set_ylabel('silhouette Score')
fig.tight_layout()
plt.show()
```

The graph above is extremely noisy, so I'm going to run many trials and make a boxplot version of this graph to see if I can reduce the noise.

```python
def silhouette_analysis(dataset, range_n_clusters):
    from sklearn.cluster import KMeans
    from sklearn.metrics import silhouette_samples, silhouette_score
    
    silhouette_avg_list = []

    for n_clusters in range_n_clusters:
        assert(n_clusters>=2)
        clusterer = KMeans(n_clusters=n_clusters)
        cluster_labels = clusterer.fit_predict(dataset)

        silhouette_avg = silhouette_score(dataset, cluster_labels)
        silhouette_avg_list.append((n_clusters, silhouette_avg))
        
    return silhouette_avg_list

def silhouette_plot(dataset, range_n_clusters=None, num_trials=5):
    
    if range_n_clusters is None:
        range_n_clusters = range(2,5)
    results = []
    for i in xrange(num_trials):
        results.append(silhouette_analysis(dataset, range_n_clusters))
    
    scores = []
    best_avg = 0
    best_k = None
    for r in zip(*results):
        for score_list in r:
            scores.append(score_list)
        avg = np.mean(zip(*r)[1])
        if avg > best_avg:
            best_avg = avg
            best_k = score_list[0]
    scores = zip(*scores)
    
    plt.figure()
    sns.boxplot(x=scores[0], y=scores[1])
    ax = plt.gca()
    ax.set_title('Bootstrapped Silhouette Analysis (%d trials)'%num_trials)
    ax.set_ylabel('silhouette score')
    ax.set_xlabel('k value')
    
    return best_k, best_avg
silhouette_plot(dataset, range(2,35),20)
```

###Preprocessing
####Naive Version

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(min_df=1, stop_words='english')
biographies = text_data['biographies']
X = vectorizer.fit_transform(biographies)
num_samples, num_features = X.shape
print "#samples: %d, #features: %d"%(num_samples, num_features)
```

```python
import scipy as sp
def dist_norm(v1, v2):
    v1_normalized = v1/sp.linalg.norm(v1)
    v2_normalized = v2/sp.linalg.norm(v2)
    delta = v1_normalized - v2_normalized
    return sp.linalg.norm(delta)

def find_euclidean_closest(new_bio, dist_func, X, corpus=biographies):
    
    new_bio_vec = vectorizer.transform([new_bio])
    new_bio_vec = new_bio_vec.toarray()
    
    best_doc = None
    best_dist = np.inf
    best_i = None
    for i, bio in enumerate(biographies):
        if bio == new_bio:
            continue
        bio_vec = X.getrow(i).toarray()
        d =  dist_func(bio_vec, new_bio_vec)
#         print '===Post %d with dist=%.2f: %s'%(i, d, bio[:100])
        if d < best_dist:
            best_dist = d
            best_i = i
        
    print("Best bio is %d with dist=%.2f"%(best_i, best_dist))
    print(text_data.ix[best_i]['track_name'])

find_euclidean_closest("this is a test.  I'm just writing a bunch of random stuff to see if this thing works at all.  My suspicion is that it doesn't",
                       dist_norm, X)
```

####Including Stemming

```python
import nltk.stem
english_stemmer = nltk.stem.SnowballStemmer('english')
class StemmedCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(StemmedCountVectorizer, self).build_analyzer()
        return lambda doc: (english_stemmer.stem(w) for w in analyzer(doc))
    
vectorizer = StemmedCountVectorizer(min_df=1, stop_words='english')
X = vectorizer.fit_transform(biographies)
num_samples, num_features = X.shape
print "#samples: %d, #features: %d"%(num_samples, num_features)
find_euclidean_closest("this is a test.  I'm just writing a bunch of random stuff to see if this thing works at all.  My suspicion is that it doesn't", 
                        dist_norm, X)
```

####TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk.stem
english_stemmer = nltk.stem.SnowballStemmer('english')
class StemmedTfidfVectorizer(TfidfVectorizer):
    def build_analyzer(self):
        analyzer = super(TfidfVectorizer, self).build_analyzer()
        return lambda doc: (
                english_stemmer.stem(w) for w in analyzer(doc))
vectorizer = StemmedTfidfVectorizer(min_df=1,
                stop_words='english', decode_error='ignore')

X = vectorizer.fit_transform(biographies)
num_samples, num_features = X.shape
print "#samples: %d, #features: %d"% (num_samples, num_features)
```

```python
vectorizer = StemmedTfidfVectorizer(
        min_df=10, max_df=0.5,
        stop_words='english',
        decode_error='ignore')
X = vectorizer.fit_transform(biographies)
num_samples, num_features = X.shape
print "#samples: %d, #features: %d"% (num_samples, num_features)
```
